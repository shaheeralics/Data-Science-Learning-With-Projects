        1. Introduction
    </h1>
    <p style="font-size:16px; font-family:Georgia, serif; line-height: 1.5em; color:#333;">
        Pakistan Sign Language (PSL) is essential for communication within the deaf community in Pakistan. This project facilitates interaction between the deaf and hearing individuals through an automated PSL interpreter that converts PSL signs into text and speech using skeletal tracking and machine learning. The system provides a cost-effective and accessible tool, supporting interactive PSL learning.
    </p>
    <p style="font-size:16px; font-family:Georgia, serif; line-height: 1.5em; color:#333;">
        The PSL dataset developed for this project includes over 2,000 images for 37 Urdu alphabet signs and 700 images for 12 Urdu words, annotated with Urdu labels. Using OpenPose, the dataset captures skeletal key points, normalized for accurate sign recognition. The model achieved high accuracy, with the alphabet recognition reaching 99.4% and word recognition 99.2% on test data, proving effective in real-time PSL interpretation.
    </p>
