        Outlier Handling Techniques and Dataset Analysis
    </h1>
    <p style="font-size:16px; font-family:Georgia, serif; line-height: 1.5em; text-indent: 20px; color:#333;">
        In this section, we will explore four different techniques for handling outliers and create four distinct datasets based on these techniques. We will also retain the original dataset without any outlier handling. The five datasets will be analyzed to assess the impact of these outlier handling techniques on model performance.
    </p>
    <p style="font-size:16px; font-family:Georgia, serif; line-height: 1.5em; text-indent: 20px; color:#333;">
        The four outlier handling techniques we will use are:
    </p>
    <ul style="font-size:16px; font-family:Georgia, serif; color:#333;">
        <li><strong>Truncation:</strong> This technique involves limiting the extreme values of a dataset to a specified range.</li>
        <li><strong>Winsorizing:</strong> This method replaces extreme values with the nearest value within a specified percentile range.</li>
        <li><strong>Z-Score Method:</strong> This technique identifies outliers based on the number of standard deviations away from the mean.</li>
        <li><strong>Modified Z-Score:</strong> This method uses a robust measure of variability to detect outliers, providing a more resilient approach to handling data with extreme values.</li>
    </ul>
    <p style="font-size:16px; font-family:Georgia, serif; line-height: 1.5em; text-indent: 20px; color:#333;">
        For each of these techniques, we will create a separate dataset. Additionally, the original dataset (without any outlier handling) will serve as our fifth dataset. We will then analyze all five datasets using various algorithms to calculate performance metrics such as R² score, RMSE (Root Mean Squared Error), and RMAE (Root Mean Absolute Error). This comprehensive analysis will help us understand how different outlier handling methods affect model performance and determine the most effective approach for this dataset.
    </p>
</div>


```python
def truncate_outliers(df, feature, lower_percentile=0.01, upper_percentile=0.99):
    lower_bound = df[feature].quantile(lower_percentile)
    upper_bound = df[feature].quantile(upper_percentile)
    df[feature] = df[feature].clip(lower=lower_bound, upper=upper_bound)

# Create a copy of the original DataFrame
truncated_laptop = laptop.copy()

# Apply truncation to all features
for feature in feature_names:
    truncate_outliers(truncated_laptop, feature)
truncated_laptop.shape
```

```python
from scipy.stats import mstats

def winsorize_feature(df, feature, limits):
    df[feature] = mstats.winsorize(df[feature], limits=limits)

# Create a copy of the original DataFrame
winsorized_laptop = laptop.copy()

# Apply winsorization to all features (e.g., 1% on each side)
for feature in feature_names:
    winsorize_feature(winsorized_laptop, feature, limits=[0.01, 0.01])
winsorized_laptop.shape
```

```python
from scipy import stats

def z_score_outlier_removal(df, feature, threshold=3):
    z_scores = stats.zscore(df[feature])
    return df[(z_scores < threshold)]

# Create a copy of the original DataFrame
z_score_filtered_laptop = laptop.copy()

# Apply Z-score outlier removal to all features
for feature in feature_names:
    z_score_filtered_laptop = z_score_outlier_removal(z_score_filtered_laptop, feature)
z_score_filtered_laptop.shape
```

```python
def modified_z_score_outlier_removal(df, feature, threshold=3.5):
    median = df[feature].median()
    mad = (df[feature] - median).abs().mean()  # Compute MAD manually
    modified_z_scores = 0.6745 * (df[feature] - median) / mad
    return df[(modified_z_scores.abs() < threshold)]

# Create a copy of the original DataFrame
modified_z_score_filtered_laptop = laptop.copy()

# Apply Modified Z-score outlier removal to all features
for feature in feature_names:
    modified_z_score_filtered_laptop = modified_z_score_outlier_removal(modified_z_score_filtered_laptop, feature)
modified_z_score_filtered_laptop.shape
```

```python

def plot_correlation_heatmap_seaborn(df, feature_names, title):
    """
    Create and display a correlation heatmap for the given dataset using Seaborn.

    Parameters:
    - df: pandas DataFrame containing the dataset.
    - feature_names: list of feature names to include in the correlation matrix.
    - title: string, the title of the heatmap.
    """
    # Calculate the correlation matrix
    correlation_matrix = df[feature_names].corr()

    # Create the heatmap
    plt.figure(figsize=(12, 10))
    sns.heatmap(
        correlation_matrix,
        annot=True,
        cmap='viridis',
        vmin=-1,
        vmax=1,
        center=0,
        linewidths=0.5,
        linecolor='gray'
    )

    # Set title and labels
    plt.title(title, size=16)
    plt.xticks(ticks=range(len(feature_names)), labels=feature_names, rotation=45, ha='center')
    plt.yticks(ticks=range(len(feature_names)), labels=feature_names, rotation=0, va='center')

    # Show the plot
    plt.show()
# List of datasets
datasets = {
    'Orignal_laptop':laptop,
    'Truncated Laptop': truncated_laptop,
    'Winsorized Laptop': winsorized_laptop,
    'Z-Score Filtered Laptop': z_score_filtered_laptop,
    'Modified Z-Score Filtered Laptop': modified_z_score_filtered_laptop
}

# Feature names
feature_names = ['Rating', 'Core_per_processor', 'Threads', 'RAM_GB', 'Storage_capacity_GB', 
                 'Display_size_inches', 'Horizontal_pixel', 'Vertical_pixel', 'ppi']

# Plot heatmaps for each dataset
for name, data in datasets.items():
    plot_correlation_heatmap_seaborn(data, feature_names, f'Correlation Heatmap ({name})')
```

```python

datasets = {
    'Truncated': truncated_laptop,
    'Winsorized': winsorized_laptop,
    'Z-Score Filtered': z_score_filtered_laptop,
    'Modified Z-Score Filtered': modified_z_score_filtered_laptop,
    'Original': laptop
}

# Function to evaluate models
def evaluate_models(X_train, X_test, y_train, y_test):
    # Initialize models
    models = {
        'XGBoost': XGBRegressor(),
        'SGDRegressor': SGDRegressor()
    }
    
    # Store results
    results = {}

    for name, model in models.items():
        # Train the model
        model.fit(X_train, y_train)
        
        # Make predictions
        y_pred = model.predict(X_test)
        
        # Calculate metrics
        r2 = r2_score(y_test, y_pred)
        rmse = mean_squared_error(y_test, y_pred, squared=False)
        mae = mean_absolute_error(y_test, y_pred)
        
        results[name] = {'R2 Score': r2, 'RMSE': rmse, 'MAE': mae}
    
    return results

# Define feature columns and target column
target_column = 'Price'
feature_columns = [col for col in laptop.columns if col != target_column]

# Results storage
overall_results = {}

# Process each dataset
for name, df in datasets.items():
    # Prepare features and target
    X = df[feature_columns]
    y = df[target_column]
    
    # One-hot encoding for categorical columns
    X_encoded = pd.get_dummies(X, drop_first=True)
    
    # Standard scaling
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_encoded)
    
    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
    
    # Evaluate models
    results = evaluate_models(X_train, X_test, y_train, y_test)
    overall_results[name] = results

# Print results
for dataset_name, metrics in overall_results.items():
    print(f"Results for {dataset_name}:")
    for model_name, scores in metrics.items():
        print(f"  {model_name}:")
        for metric_name, score in scores.items():
            print(f"    {metric_name}: {score:.4f}")
```

```python
 
# Define your datasets
datasets = {
    'Truncated': truncated_laptop,
    'Winsorized': winsorized_laptop,
    'Z-Score Filtered': z_score_filtered_laptop,
    'Modified Z-Score Filtered': modified_z_score_filtered_laptop,
    'Original': laptop
}

# Function to apply frequency encoding
def frequency_encode(df, feature):
    freq = df[feature].value_counts()
    df[feature + '_encoded'] = df[feature].map(freq)
    return df

# Function to evaluate models
def evaluate_models(X_train, X_test, y_train, y_test):
    # Initialize models
    models = {
        'Linear Regression': LinearRegression(),
        'XGBoost': XGBRegressor(),
        'SGDRegressor': SGDRegressor()
    }
    
    # Store results
    results = {}

    for name, model in models.items():
        # Train the model
        model.fit(X_train, y_train)
        
        # Make predictions
        y_pred = model.predict(X_test)
        
        # Calculate metrics
        r2 = r2_score(y_test, y_pred)
        rmse = mean_squared_error(y_test, y_pred, squared=False)
        mae = mean_absolute_error(y_test, y_pred)
        
        results[name] = {'R2 Score': r2, 'RMSE': rmse, 'MAE': mae}
    
    return results

# Define feature columns and target column
target_column = 'Price'
feature_columns = [col for col in laptop.columns if col != target_column]

# Results storage
overall_results = {}

# Process each dataset
for name, df in datasets.items():
    # Prepare features and target
    X = df[feature_columns].copy()
    y = df[target_column]
    
    # Apply frequency encoding to categorical columns
    for col in X.select_dtypes(include=['object']).columns:
        X = frequency_encode(X, col)
        X = X.drop(columns=[col])  # Drop the original categorical column
    
    # Drop any columns that were created during encoding
    X = X.loc[:, ~X.columns.str.endswith('_encoded')]
    
    # Standard scaling
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
    
    # Evaluate models
    results = evaluate_models(X_train, X_test, y_train, y_test)
    overall_results[name] = results

# Print results
for dataset_name, metrics in overall_results.items():
    print(f"Results for {dataset_name}:")
    for model_name, scores in metrics.items():
        print(f"  {model_name}:")
        for metric_name, score in scores.items():
            print(f"    {metric_name}: {score:.4f}")
```
